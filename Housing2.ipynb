```python
# import libraries
import geopandas as gpd
from shapely.geometry import Point, Polygon
import matplotlib.pyplot as plt
# import street map
street_map = gpd.read_file('geo_export_68a1a97b-4ca2-4b22-b7b5-a43777040f48.shp')
```

```python
# designate coordinate system
crs = {'init': 'epsg:4326'}
# zip x and y coordinates into single feature
geometry = [Point(xy) for xy in zip(dfprop['Longitude'], dfprop['Latitude'])]
# create GeoPandas dataframe
geo_df = gpd.GeoDataFrame(dfprop, crs = crs, geometry = geometry)
```

```python
# create figure and axes, assign to subplot
sns.set_style('white')
fig, ax = plt.subplots(figsize=(15,15))
# add .shp mapfile to axes
street_map.plot(ax=ax, alpha=0.4,color='grey')
# add geodataframe to axes
# assign ‘price’ variable to represent coordinates on graph
# add legend
# make datapoints transparent using alpha
# assign size of points using markersize
geo_df.plot(column='Assessed Value',ax=ax,alpha=0.5, legend=True,markersize=10,cmap="viridis")
# add title to graph
plt.title('Assessed Value of Porperties in Edmonton', fontsize=15,fontweight='bold')
# set latitiude and longitude boundaries for map display

# show map
plt.show()
```

```python
from statsmodels.formula.api import ols
```

```python
fit = ols('Assessed_Value ~ C(Ward) + C(Garage)', data=dfprop).fit() 
fit.summary()
```

```python
import statsmodels.formula.api as smf
```

```python

model = smf.ols('Assessed_Value  ~ C(Ward) + C(Garage)', data=dfprop)
results = model.fit()
results.summary()
```

```python
endog = pd.DataFrame(model.endog, columns=[model.endog_names])
exog = pd.DataFrame(model.exog, columns=model.exog_names)
```

```python
actual = endog['Assessed_Value']
baseline = actual.mean()
baseline
```

```python
predict = (results.predict() >= 0.5)
true_pos = predict * actual
true_neg = (1 - predict) * (1 - actual)
sum(true_pos), sum(true_neg)
```

```python
acc = (sum(true_pos) + sum(true_neg)) / len(actual)
acc
```
